```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target == 1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
df0.head()

plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', label='setosa')

df1.head()

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', label='versicolor')

df2.head()
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='green', label='virginica')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
X = df.drop(['target', 'flower_name'], axis='columns')
y = df.target
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.2)
```

```python
model = SVC()
```

```python

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
